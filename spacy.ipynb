```python
text = '''
Have you heard about the latest sale going on at our store? You can save up to 50% off on selected items, including clothing,
accessories, and home decor. This is a limited-time offer, so make sure to take advantage of it before it's too late! If you
have any questions about the sale or the products we offer, please feel free to email us at customerservice@store.com. 
We're always happy to help. And if you want to stay updated on our future sales and promotions, make sure to subscribe 
to our email list at deals@store.com. You can also follow us on social media and share your favorite finds with us at 
social@store.com. We love hearing from our customers and appreciate your continued support.
'''
text
```

```python
doc = nlp(text)
emails = []
for token in doc:
    if token.like_email:
        emails.append(token.text)
emails
```

```python
nlp = spacy.blank('hi')

text = '''
हिंदी भाषा एक बहुत ही सुंदर और समृद्ध भाषा है। यह भारत की आधिकारिक भाषा होने के साथ-साथ दुनिया के कई अन्य देशों में भी बोली जाती है। हिंदी भाषा में कई 
महान शायर, कवि और लेखक हैं जो इस भाषा की 5000₹ सुंदरता और समृद्धता को निखारते हुए अपने शब्दों में आत्मा का बोध कराते हैं। हिंदी भाषा विश्व की सबसे प्रचलित 
भाषाओं में से एक है, जो भारतीय सभ्यता, विरासत और अनुभव की विस्तृत और गहरी जानकारी को दर्शाती है।
'''
doc = nlp(text)
for token in doc:
    print(token, token.is_currency)
```
