```python
# spacy is object-oriented whereas, nltk is string processing library
import spacy 
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
#doc is object & sents is its property hence a object-oriented
doc= nlp("Dr. Strange loves pav bhaji of mumbai. Hulk loves chat of delhi")
for sentence in doc.sents:
    print(sentence)
```

```python
for sentence in doc.sents:
    for words in sentence:
        print(words)
```

```python
# spacy is like mobile phone camera, nltk is more like of pro mode in camera 
# which have various setting to get a best photo, but spacy have auto-adjusted &
# less-feauteristic approach to solve the problem or mainly tokenizing the text
#
from nltk.tokenize import sent_tokenize
sent_tokenize("Dr. Strange loves pav bhaji of mumbai. Hulk loves chat of delhi")
```

```python
from nltk.tokenize import word_tokenize
word_tokenize("Dr. Strange loves pav bhaji of mumbai. Hulk loves chat of delhi")
```
