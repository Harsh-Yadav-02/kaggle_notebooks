```python
import spacy
```

```python
nlp = spacy.blank("en")

doc = nlp("Captain america ate 100$ of samosa, Then he said I can do this all day.")

for token in doc:
    print(token)
```

```python
doc = nlp("Captain america ate 100$ of samosa, Then he said I can do this all day.")

for token in doc:
    print(token, " | ", token.pos_, " | ", token.lemma_)
```
